## Codificación de la imagen

```python
# Usamos el módulo de tratamiento de imágenes PIL.
from PIL import Image
from numpy import array
# Usaremos Sympy
from sympy import *
from criptoMRG import *
```

```python
imag = Image.open('./imcod.png')
imag
```

```python
# Ubicamos y obtenemos los dos puntos rojos
imagen = imag.copy()
(n,m) = imagen.size
pixels = imagen.load()

# Nótese que los dos puntos rojos se han obtenido en las esquinas superior
# izquierda e inferior derecha.
pixels[0,0],pixels[n-1,m-1]

# -X   + R = 1
# X0   + B = 11
# X0   + G = 222

## Tenemos
# B = 11
# G = 222
# X = R-1

for coeff in xrange(255):
    rvect = (coeff+1)%256
    gvect = 11
    bvect = 222
     
    if gcd(coeff, 256) == 1:
        nueva = afinimgDec(imagen, coeff, rvect, gvect, bvect)
        nueva.save("decodimages/imagen_{}.png".format(coeff), 'PNG')

# Obtenemos que la imagen 123 funciona bien, esto nos da un coeficiente 123
```

```python
coeff = 123
rvect = (coeff+1)%256
gvect = 11
bvect = 222

afinimgDec(imagen, coeff, rvect, gvect, bvect) 
```
